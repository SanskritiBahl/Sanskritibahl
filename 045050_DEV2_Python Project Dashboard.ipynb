```python
#### Author : Sanskriti Bahl
###  Term   : 2
### Project 2 :  Preparing Interactive Dashboard

## For preparing dashboard, dash has been used since with streamlit I was unable to exceute the project properly

### Financial dashboard has been created where different the variations of the stock prices of various stocks can be observed through out the day
### This kind of dashboard help the traders and investors to analyze and understand the volaality of various stocks. This thus help in better understanding
### of volatile and stable stock. This type of understanding helps in the deciosn making of risk-averse and risk-taker investors.
### Objective : The objective of the report is analyze the intra-day trading of various stocks based on corresponding closing prices. This analysis helps
### helps the investor in understanding the volatality of the stock.

### Data source  : yfinance
### Libaray used for creating dashboard : Dash

!pip install dash
import datetime
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import yfinance as yf
import plotly.graph_objs as go
import pandas as pd
import threading

# Function to load real-time stock data
def load_real_time_data(stock_symbol):
    stock_data = yf.download(stock_symbol, period="1d", interval="1m")
    return stock_data

# Initialize the Dash app
app = dash.Dash(__name__)
app.title = 'Real-Time Stock Dashboard'

# Define the layout of the app
app.layout = html.Div([
    html.H1("Real-Time Stock Dashboard"),

    html.Div([
        html.Label("Select Stock Symbol:"),
        dcc.Input(id='stock-input', value='AAPL', type='text'),
    ]),

    dcc.Tabs([
        dcc.Tab(label='Line Chart', children=[
            dcc.Graph(id='line-chart'),
        ]),

        dcc.Tab(label='Candlestick Chart', children=[
            dcc.Graph(id='candlestick-chart'),
        ]),

        dcc.Tab(label='Bar Chart', children=[
            dcc.Graph(id='bar-chart'),
        ]),

        dcc.Tab(label='Scatter Plot', children=[
            dcc.Graph(id='scatter-plot'),
        ]),
    ]),

    dcc.Interval(
        id='interval-component',
        interval=1 * 60 * 1000,  # in milliseconds (update every 1 minute)
        n_intervals=0
    ),
])

# Callback to update real-time charts
@app.callback(
    [Output('line-chart', 'figure'),
     Output('candlestick-chart', 'figure'),
     Output('bar-chart', 'figure'),
     Output('scatter-plot', 'figure')],
    [Input('stock-input', 'value'),
     Input('interval-component', 'n_intervals')]
)
def update_real_time_charts(stock_symbol, n_intervals):
    stock_data = load_real_time_data(stock_symbol)

    # Line Chart
    line_chart_fig = {
        'data': [go.Scatter(x=stock_data.index, y=stock_data['Close'], mode='lines')],
        'layout': go.Layout(title=f'{stock_symbol} - Real-Time Closing Price')
    }

    # Candlestick Chart
    candlestick_fig = {
        'data': [go.Candlestick(x=stock_data.index,
                                open=stock_data['Open'],
                                high=stock_data['High'],
                                low=stock_data['Low'],
                                close=stock_data['Close'])],
        'layout': go.Layout(title=f'{stock_symbol} - Real-Time Candlestick Chart')
    }

    # Bar Chart
    bar_chart_fig = {
        'data': [go.Bar(x=stock_data.index, y=stock_data['Volume'])],
        'layout': go.Layout(title=f'{stock_symbol} - Real-Time Volume')
    }

    # Scatter Plot
    scatter_plot_fig = {
        'data': [go.Scatter(x=stock_data.index, y=stock_data['Close'], mode='markers')],
        'layout': go.Layout(title=f'{stock_symbol} - Real-Time Scatter Plot')
    }

    return line_chart_fig, candlestick_fig, bar_chart_fig, scatter_plot_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
```
